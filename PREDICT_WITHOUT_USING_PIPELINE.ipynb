```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("/content/Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
# Step 1 -> train/test/split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                random_state=42)
```

```python
si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
X_train_embarked
```

```python
# one hot encoding Sex and Embarked

# Remove 'sparse=False' since it is deprecated. If you want to ensure a dense output array, call toarray() method on output
ohe_Sex = OneHotEncoder(handle_unknown='ignore')
ohe_Embarked = OneHotEncoder(handle_unknown='ignore')

X_train_Sex = ohe_Sex.fit_transform(X_train[['Sex']]).toarray()
X_train_Embarked = ohe_Embarked.fit_transform(X_train_embarked).toarray() # X_train_Embarked should be the NumPy array from imputation

X_test_Sex = ohe_Sex.transform(X_test[['Sex']]).toarray()
X_test_Embarked = ohe_Embarked.transform(X_test_embarked).toarray()
```

```python
X_train_Embarked
```

```python
X_train.head()
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_Sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_Sex,X_test_embarked),axis=1)
```

```python
X_test_transformed.shape
```

```python
# one hot encoding Sex and Embarked
# Remove 'sparse=False' since it is deprecated. If you want to ensure a dense output array, call toarray() method on output
# The issue is the input to OneHotEncoder needs to be the NumPy array, not the original dataframe slice
ohe_Sex = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  # Use sparse=False for older scikit-learn versions
ohe_Embarked = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # Use sparse=False for older scikit-learn versions

X_train_Sex = ohe_Sex.fit_transform(X_train[['Sex']])
X_train_Embarked = ohe_Embarked.fit_transform(X_train[['Embarked']])

X_test_Sex = ohe_Sex.transform(X_test[['Sex']])
X_test_Embarked = ohe_Embarked.transform(X_test[['Embarked']])

# Concatenate transformed features
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])

# Assuming X_train_age and X_test_age are already numerical
X_train_transformed = np.concatenate((X_train_rem, X_train_age, X_train_Sex, X_train_Embarked), axis=1)
X_test_transformed = np.concatenate((X_test_rem, X_test_age, X_test_Sex, X_test_Embarked), axis=1)

# Now fit the model
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed, y_train)
```

```python

y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
