```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

2.WORKING WITH YEAR

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

```python
#EXTREACT MONTH
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#Extract Quarter
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#Extract Semester
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
# Calculate the difference in years and months separately
months_passed = (today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)

# Optional: Adjust for the day of the month if more precision is needed
# For a rough estimate, the above is sufficient.
# If you need to consider if a full month has passed based on the day:
# months_passed = months_passed - (today.day < date['date'].dt.day).astype(int)

np.round(months_passed, 0)
```
